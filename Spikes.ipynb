```python
from bs4 import BeautifulSoup, NavigableString, Tag
```

```python
def do_shit(soup):
    start = soup.body.span
    for i in soup.find_all("span"):
        if i.next_sibling.string.find("START") != -1:
            start = i
            start2 = i   

    premoves = []
    start2 = start.next_sibling

    while True :
        if start2.previous_sibling is None:
            break
        if isinstance(start2.previous_sibling, NavigableString):
            pass
        if isinstance(start2.previous_sibling, Tag):
            if start2.previous_sibling != '\n':
                if start2.previous_sibling.get("class")[0].find("214") != -1:
                    premoves.append(start2.previous_sibling.string.split(' ')[1] + '\n')
        if len(premoves) == 2:
            break
        start2 = start2.previous_sibling
    

    if len(premoves) == 2:
        premoves = [premoves[1], premoves[0]]

    start2 = start
    hp_before_start = -1
    while True :
        if start2.previous_sibling is None:
            break
        if isinstance(start2.previous_sibling, NavigableString):
            pass
        if isinstance(start2.previous_sibling, Tag):
            if start2.previous_sibling != '\n':
                if start2.previous_sibling.get("class")[0].find("199") != -1:
                    hp_before_start = start2.previous_sibling.string
                    break
        start2 = start2.previous_sibling

    moves = []
    start2 = start
    while start2.string.find("END") == -1:
        if start2 is None:
            break
        if isinstance(start2, NavigableString):
            pass
        if isinstance(start2, Tag):
            if start2 != '\n':
                if start2.get("class")[0].find("214") != -1:
                    moves.append(start2.string.split(' ')[1] + '\n')
        start2 = start2.next_sibling

    hp_after_end = -1
    while True :
        if start2 is None:
            break
        if isinstance(start2, NavigableString):
            pass
        if isinstance(start2, Tag):
            if start2 != '\n':
                if start2.get("class")[0].find("199") != -1:
                    hp_after_end = start2.string
                    break
        start2 = start2.next_sibling
    
    
    
    victim = ''.join([soup.pre.contents[0].split('\n')[i] for i in [1]])
    victim = ''.join([victim.split(' ')[i] for i in [-4,-2,-1]])
    
    killer = ''.join([soup.pre.contents[0].split('\n')[i] for i in [2]])
    killer = ''.join([killer.split(' ')[i] for i in [-4,-2,-1]])
    
    return [killer, victim, premoves, hp_before_start, moves, hp_after_end]
```

```python
import xlwt

def output(filename, book, sh, variables, padding, source):
    killer_desc = 'Убийца'
    victim_desc = 'Жертва'
    premoves_desc = 'Действия до START'
    hp_b_s_desc = 'Хп было'
    moves_desc = 'Действия'
    hp_a_e_desc = 'Хп стало'
    desc = [killer_desc, victim_desc, premoves_desc, hp_b_s_desc, moves_desc, hp_a_e_desc]
    
    sh.write(padding, 0, source)

    for n, (v_desc, v) in enumerate(zip(desc, variables)):
        sh.write(n + 1 + padding, 0, v_desc)
        sh.write(n + 1 + padding, 1, v)

    book.save(filename)
```

```python
from os import listdir
from os.path import isfile, join
mypath = "C:/Users/rubal/OneDrive/Desktop/webm/Spikes/Jet"
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]

book = xlwt.Workbook(encoding="utf-8")
sheet = book.add_sheet("Sheet 1")

j = 0

for f,i in zip(onlyfiles, range(len(onlyfiles))):
    print(f)
    file = mypath + "/" + f
    print(file)
    fp = open(file, encoding='utf-8')
    soup = BeautifulSoup(fp, "html.parser")
    
    if ''.join([''.join([soup.pre.contents[0].split('\n')[i] for i in [1]]).split(' ')[i] for i in [-4, -1]]) == "104":
        j = j + 1
        print("continued")
        continue
    
    v = do_shit(soup)
    output("jet_result.xls", book, sheet, v, 9*(i - j) + 1, f)

    print(j)
    print(len(onlyfiles))
```
